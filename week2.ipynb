membuat bentuk geometri 2D dari opencv diawali dengan pembuatan black image dengan tipe data int16

```python
black_wiki = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_wiki)
```

```python
#perhatikan koordinat titik2 pt1 dan pt2
cv.rectangle(black_wiki,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=10)
plt.imshow(black_wiki)
```

```python
cv.rectangle(black_wiki,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=15)
plt.imshow(black_wiki)
```

```python
cv.circle(black_wiki,center=(100,100),radius=50,color=(255,0,0),thickness=8)
plt.imshow(black_wiki)
```

```python
cv.line(black_wiki,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_wiki)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_wiki,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255.255,10),thickness=2,lineType=cv.LINE_AA)
plt.imshow(black_wiki)
```

```python
black_wiki2=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_wiki2)
```

```python
#inisialisasi numpy array dengan tipe data int32
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
vertices
```

```python
pts = vertices.reshape((-1,1,2)) #nilai 2 menunjukkan bahwa tiap titik dibuat 3 chanel yang mewakili RGB
pts
```

```python

cv.polylines(black_wiki2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_wiki2)
```

```python
wiki = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/kitten0.jpg')
R = wiki.copy();
R[:,:,1]=0
R[:,:,2]=0
print('Red Shape =',R.shape)

g = wiki.copy();
g[:,:,0]=0
g[:,:,2]=0
print('green Shape =',g.shape)

b = wiki.copy();
b[:,:,1]=0
b[:,:,2]=0
print('blue Shape =',b.shape)

fig, axes = plt.subplot(2,2, figsize=(10,10))
ax = axes.ravel()

ax[0].imshow(wiki)
ax[0].set_title("original")
ax[1].imshow(R)
ax[1].set_title("redchanel")
ax[2].imshow(g)
ax[2].set_title("greenchanel")
ax[3].imshow(b)
ax[3].set_title("bluechanel")
fig.tight_layout()
plt.show
```
